```python
from datasets import load_dataset

dataset = load_dataset("traversaal-ai-hackathon/hotel_datasets")
```

```python
import chromadb
from chromadb.utils import embedding_functions
chroma_client = chromadb.Client()
```

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
```

```python
collection = chroma_client.delete_collection(name="my_collection")
```

```python
collection = chroma_client.create_collection(name="my_collection", embedding_function=sentence_transformer_ef)
```

```python
list = df.values.tolist()
```

```python
print(list[11][1])
```

```python
documents = []
metadatas = []
ids = []
id = 0
for i, line in enumerate(list):
  documents.append(line[0])
  metadatas.append({"hotel description: ": line[0]})
  # print(line[1])
  ids.append(str(i+1))

  


  
```

```python
collection.add(
    documents = documents,
    metadatas = metadatas,
    ids = ids
)
```

```python
results = collection.query(
    query_texts=["I want to book a hotel in Paris with a senic view and a pool"],
    n_results=10
)
results
```
